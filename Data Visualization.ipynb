```python
A = A.drop(labels=["id","Make","Model"],axis=1)
```

# Missing data treatment

```python
from important_tool import replacer
replacer(A)
```

```python
cat = []
con = []
for i in A.columns:
    if(A[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
    
```

# Create distribution plot for all continuous columns

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.subplot(3,3,1)
sb.distplot(A['Min.Price'])
plt.subplot(3,3,2)
sb.distplot(A.Price)
plt.subplot(3,3,3)
sb.distplot(A['Max.Price'])
plt.subplot(3,3,4)
sb.distplot(A['MPG.city'])
plt.subplot(3,3,5)
sb.distplot(A['MPG.highway'])
plt.subplot(3,3,6)
sb.distplot(A.EngineSize)
```

```python
nrows = len(con)/3
nrows
```

```python
plt.figure(figsize=(20,20))
w = 1
for i in con:
    plt.subplot(nrows,3,w)
    sb.distplot(A[i])
    w = w + 1
```
